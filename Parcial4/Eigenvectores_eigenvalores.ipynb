# Eigenvectores y Eigenvalores

## ¿Qué son?

Dado una **matriz cuadrada** $A$, un **eigenvector** $\vec{v}$ es un vector **no nulo** tal que:

$$
A \vec{v} = \lambda \vec{v}
$$

- $\vec{v}$ es el **eigenvector**
- $\lambda$ es el **eigenvalor**
- $A$ es una matriz cuadrada de dimensión $n \times n$

---

## Interpretación geométrica

Un **eigenvector** no cambia de dirección cuando se le aplica la transformación $A$. Solo cambia su **magnitud** (escalado por $\lambda$).

- Si $\lambda > 1$: el vector se **alarga**
- Si $0 < \lambda < 1$: el vector se **encoge**
- Si $\lambda < 0$: además de escalarse, **cambia de sentido**

---

## Obteniendo los eigenvalores

Partimos de la definición de eigenvector:

$$
A \vec{v} = \lambda \vec{v}
$$

Multiplicamos $\lambda$ por la matriz identidad $I$ (para poder restarlo de $A$):

$$
A \vec{v} = \lambda I \vec{v}
$$

Reorganizamos términos:

$$
A \vec{v} - \lambda I \vec{v} = 0 \Rightarrow (A - \lambda I)\vec{v} = 0
$$

Este es un **sistema homogéneo**. Queremos una **solución no trivial** (es decir, $\vec{v} \ne 0$).  
Para que eso ocurra, la matriz $A - \lambda I$ debe ser **no invertible**, es decir, su determinante debe ser cero:

$$
\det(A - \lambda I) = 0
$$

Esta ecuación se llama el **polinomio característico**. Sus soluciones $\lambda$ son los **eigenvalores** de $A$.

---

## Ejemplo numérico: paso a paso

Sea la matriz:

$$
A = \begin{bmatrix} 2 & 0 \\ 0 & 3 \end{bmatrix}
$$

### 1. Encontrar los eigenvalores

Calculamos:

$$
\det(A - \lambda I) = 
\det\begin{bmatrix}
2 - \lambda & 0 \\
0 & 3 - \lambda
\end{bmatrix}
= (2 - \lambda)(3 - \lambda)
$$

Igualamos a cero:

$$
(2 - \lambda)(3 - \lambda) = 0 \Rightarrow \lambda_1 = 2, \quad \lambda_2 = 3
$$

---

### 2. Encontrar los eigenvectores

#### Para $$\lambda = 2$$:

$$
A - 2I = 
\begin{bmatrix}
0 & 0 \\
0 & 1
\end{bmatrix}
$$

Resolviendo $$(A - 2I)\vec{v} = 0$$:

$$
\begin{bmatrix}
0 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
0 \\
y
\end{bmatrix}
= 
\begin{bmatrix}
0 \\
0
\end{bmatrix}
\Rightarrow y = 0
$$

Entonces:

$$
\vec{v}_1 = \begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

---

#### Para $$\lambda = 3$$:

$$
A - 3I = 
\begin{bmatrix}
-1 & 0 \\
0 & 0
\end{bmatrix}
$$

$$
\begin{bmatrix}
-1 & 0 \\
0 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
-x \\
0
\end{bmatrix}
= 
\begin{bmatrix}
0 \\
0
\end{bmatrix}
\Rightarrow x = 0
$$

Entonces:

$$
\vec{v}_2 = \begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

---

### 3. Verificación de la igualdad $A \vec{v} = \lambda \vec{v}$

#### Para $$\lambda_1 = 2$$ y $$\vec{v}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$$:

$$
A \vec{v}_1 =
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
\begin{bmatrix}
2 \\
0
\end{bmatrix}
= 2 \cdot 
\begin{bmatrix}
1 \\
0
\end{bmatrix}
= \lambda_1 \vec{v}_1
$$

✔️ Se cumple.

#### Para $$\lambda_2 = 3$$ y $$\vec{v}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$$:

$$
A \vec{v}_2 =
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}
=
\begin{bmatrix}
0 \\
3
\end{bmatrix}
= 3 \cdot 
\begin{bmatrix}
0 \\
1
\end{bmatrix}
= \lambda_2 \vec{v}_2
$$

✔️ También se cumple.

---

## Conclusión

- Los eigenvalores de $A$ son $\lambda_1 = 2$ y $\lambda_2 = 3$  
- Los eigenvectores correspondientes son $\vec{v}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ y $\vec{v}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$
- Se verificó que $$A \vec{v} = \lambda \vec{v}$$ se cumple en ambos casos
